## ðŸŽ² Linear probability models (LPMs)

Fit a linear regression model to the spam data with the `lm` function.

```python
X = emails[['char_dollar', 'credit', 'money', 're']]
y = emails['is_spam']

X = sm.add_constant(X)

mdl = sm.OLS(y, X).fit()
print(mdl.summary())
```

## ðŸŽ° Odds functions

Write two functions:
- A function to convert probabilities to odds.
- A function to convert odds to probabilities

Test your functions by making sure that 2:1 odds returns a 2/3 probability, and vice versa.

Finally, suppose my probability of winning is 60%. If I double my odds of winning, what is my new probability of winning?

```python
def prob_to_odds(p):
    return p / (1 - p)

def odds_to_prob(odds):
    return odds / (1 + odds)

# Test
test1 = odds_to_prob(2)
test2 = prob_to_odds(2/3)
test1, test2
```

```python
prob = 0.60
odds = prob_to_odds(prob)
new_odds = 2 * odds
new_prob = odds_to_prob(new_odds)
print(f"new probability : {new_prob:.3f}")
```

## ðŸª™ Fitting a logistic regression model

We can fit a logistic regression model with the same covariates as above with the following code:

```python
import statsmodels.formula.api as smf

logistic_mdl = smf.glm(
    "is_spam ~ char_dollar + credit + money + re",
    data=emails,
    family=sm.families.Binomial()
).fit()

print(logistic_mdl.summary())
```

```python
logistic_model = sm.GLM(y, X, family=sm.families.Binomial()).fit()
print(logistic_model.summary())
```

Interpret the intercept and `money` coefficients for the logistic regression model three different ways:
1. On the log odds scale
2. On the odds scale (by exponentiating the coefficients)
3. On the probability scale (using either the odds functions you wrote, or the divide by 4 trick).

Tip: Use the `coef` function to extract coefficients from the model.

**1. Log odd scale**

```python
#coefficients
coef = logistic_model.params
intercept = coef['const']
money_coef = coef['money']

print(f"Intercept: {intercept:.3f}")
print(f"Money coefficient: {money_coef:.3f}")
```

**2. Odd scale**

```python
intercept_odds = np.exp(intercept)
money_odds = np.exp(money_coef)

print(f"Intercept: {intercept_odds:.3f}")
print(f"Money coefficient: {money_odds:.3f}")
```

**3. Probability scale**

```python
intercept_prob = intercept_odds / (1 + intercept_odds)
money_prob_change = money_coef / 4

print(f"Intercept: {intercept_prob:.3f}")
print(f"Money coefficient: {money_prob_change:.3f}")
```
